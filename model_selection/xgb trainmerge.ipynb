```python
import pandas as pd
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('C:/Users/hp/Downloads/train_merge (2).csv')
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

df['gender'] = label_encoder.fit_transform(df['gender'])
df['signup_method'] = label_encoder.fit_transform(df['signup_method'])
df['language'] = label_encoder.fit_transform(df['language'])
df['affiliate_channel'] = label_encoder.fit_transform(df['affiliate_channel'])
df['affiliate_provider'] = label_encoder.fit_transform(df['affiliate_provider'])
df['signup_app'] = label_encoder.fit_transform(df['signup_app'])
df['first_affiliate_tracked'] = label_encoder.fit_transform(df['first_affiliate_tracked'])
df['first_device_type'] = label_encoder.fit_transform(df['first_device_type'])
df['first_browser'] = label_encoder.fit_transform(df['first_browser'])
df['most_frequent_action'] = label_encoder.fit_transform(df['most_frequent_action'])
df['country_destination'] = label_encoder.fit_transform(df['country_destination'])
df['most_frequent_action'] = label_encoder.fit_transform(df['most_frequent_action'])
df['most_frequent_action_type'] = label_encoder.fit_transform(df['most_frequent_action_type'])
df['most_frequent_action_detail'] = label_encoder.fit_transform(df['most_frequent_action_detail'])
df['most_frequent_device_type'] = label_encoder.fit_transform(df['most_frequent_device_type'])

df
```

```python
df= df.drop(columns=['date_account_created','timestamp_first_active','date_first_booking'],axis=1)
df
```

```python
y = df['country_destination']

X = df.drop(columns=['country_destination'])
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
y_train
```

```python
import xgboost as xgb
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
```

```python
# Initialize XGBClassifier
model = XGBClassifier(
    objective='multi:softmax', 
    num_class=12,               
    n_jobs=-1 
)

param_grid = {
    'max_depth': [3, 6, 9],
    'learning_rate': [0.01, 0.1, 0.2],
    'n_estimators': [100, 200, 300],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0],
}

grid_search = GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    scoring='accuracy',  
    cv=3  
)

grid_search.fit(X_train, y_train)

best_model = grid_search.best_estimator_

y_pred = best_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Best Model Accuracy: {accuracy:.2f}")
```
